# Data Visualization

```python
df.head()
```

```python
# Creating a Scatter plot chart to visualize Store Area Vs Store Sales
df.plot.scatter(x = 'Store_Area', y = 'Store_Sales', figsize=(10,6))

# label the axis
plt.xlabel('Store_Area')
plt.ylabel('Store_Sales')

# set the title
plt.title('Store Area Vs Store Sales')

# show the visual
plt.show()
```

```python
# Creating a boxplot chart to visualize Items Available, Daily Customer Count and Store Sales
plt.figure(figsize=(10, 6))
sns.boxplot(df[['Items_Available', 'Daily_Customer_Count', 'Store_Sales']])

# set the title
plt.title('Items Available, Daily Customer Count, and Store Sales')

# show the visual
plt.show()
```

```python
# Calculate correlations between variables
correlations = df.corr()
print(correlations)
```

```python
# Creating a Heatmap to visualize Correlations between the variables
plt.figure(figsize=(10, 6))
sns.heatmap(correlations, annot=True, cmap='hot')

# set the titel
plt.title('Correlation Heatmap')

# Plot the visual 
plt.show()
```

```python
# Creating a Scatter plot chart to visualize Items Available Vs Store Sales
df.plot.scatter(x = 'Items_Available', y = 'Store_Sales', figsize=(10,6))

# label the axis
plt.xlabel('Items_Available')
plt.ylabel('Store_Sales')

# set the title
plt.title('Items Available Vs Store Sales')

# show the visual
plt.show()
```

```python
# Plot a Scatter plot for Store Sales by Items Available and Store Area
plt.scatter(df['Store_Area'], df['Items_Available'], c=df['Store_Sales'])

#Add labels and title
plt.xlabel('Store Area')
plt.ylabel('Items Available')
plt.colorbar(label='Store Sales')

#Show Plot
plt.show()
```
